## Cleaning the data

```python
startup_data.isnull().sum()#Checking of null values 
```

```python
#Removing columns that have maximum null values Remarks and not required Serial no
startup_data = startup_data.drop(['Sr No','Remarks'],axis=1)
startup_data.head()
```

```python
#Making a copy of dataset 
startup_data1 = startup_data.copy()
#Rename the columns
startup_data1.columns=['Date','Startup Name','Industry Vertical','SubVertical',
                      'City Location','Investors Name','Investment Type',
                      'Amount in USD']
startup_data1

#Drop amount if Zero
startup_data1=startup_data1.drop(startup_data1[startup_data1['Amount in USD'].isnull()].index)


#Drop rows where the amount is undisclosed
startup_data1=startup_data1.drop(startup_data1[(startup_data1['Amount in USD']=='Undisclosed'  
                                               )|(startup_data1['Amount in USD']=='undisclosed' )].index)
#Remove unknown amount
startup_data1=startup_data1.drop(startup_data1[(startup_data1['Amount in USD']=='Unknown'  
                                               )|(startup_data1['Amount in USD']=='unknown' )].index)
startup_data1=startup_data1.drop(startup_data1[(startup_data1['Amount in USD']=='NA'  
                                               )|(startup_data1['Amount in USD']=='na' )].index)

#Since amount is having delimiter ',', so converting amount object type into float
startup_data1['Amount in USD'] = startup_data1['Amount in USD'].apply(lambda x: x.replace(',',''))
#Replacing '+' from some amounts
startup_data1['Amount in USD'] = startup_data1['Amount in USD'].apply(lambda x: x.replace('+',''))
# removing  invalid float data
startup_data1 = startup_data1[startup_data1['Amount in USD'].str.replace('.','',1).str.isdigit()]

#convert amount object type to float type
#Convert the data into float
startup_data1['Amount in USD'] = startup_data1['Amount in USD'].astype('float')
startup_data1['Amount in USD'].head(10)
```

```python
#columns 'Industry Vertical' and 'SubVertical' contains nan vaues
#so, filling with others
startup_data1['Industry Vertical'] = startup_data1['Industry Vertical'].fillna('Others')

startup_data1['SubVertical'] = startup_data1['SubVertical'].fillna('Others')

startup_data1['Industry Vertical'].value_counts()
for i in startup_data1['Industry Vertical']:
    print(i)
```

```python
#replacing i


#checking the City Location
#check for null values
print('Null Values in City Location are:',len(startup_data1[startup_data1['City Location'].isnull()]))

#Replace Null values with others
startup_data1['City Location'] = startup_data1['City Location'].fillna('Others')

#avoid duplicate names of cities
#format the cities names by removing other characters  than city name
startup_data1['City Location'] = startup_data1['City Location'].replace(['Bangalore','Bengaluru'],'Bangalore')


startup_data1['City Location'].head(10)
```

## Funding by cities

```python
#Cities receving max funding
fundings_count_city = pd.DataFrame(startup_data1['City Location'].value_counts().sort_values(ascending=False)[:10])
fundings_count_city.rename(columns={'City Location':'Number of Fundings by City'}, inplace=True)
print(fundings_count_city)

#Total fundings being recieved by cities
funding_total_city = pd.DataFrame(startup_data1.groupby('City Location')['Amount in USD'].sum()).sort_values(by="Amount in USD", ascending=False)[:15]
funding_total_city.rename(columns={'Amount in USD':'Total Funding by City'}, inplace=True)
funding_total_city.head()

```

### Who are the important investors in the Indian Ecosystem?
